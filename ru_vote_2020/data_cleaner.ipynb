# Transform Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('results2020.txt', sep='\t')
```

```python
data.head()
```

```python
df = data.copy()
df = df.drop(['name', 'url'], axis=1)
df = df.rename(
    {'reg': 'region',
    'Число участников голосования, включенных в список участников голосования на момент окончания голосования': 'nominal',
    'Число бюллетеней, выданных участникам голосования': 'given',
    'Число бюллетеней, содержащихся в ящиках для голосования': 'received',
    'Число недействительных бюллетеней': 'damaged',
    'ДА': 'yes', 'НЕТ': 'no'},
    axis=1)

nans = df['nominal'].isna().values
print(f'Nans in nominal (to remove): {np.sum(nans)}')
df = df[:][np.logical_not(nans)]

df['uik'] = df['uik'].str[5:].astype(int)
df[['nominal', 'given', 'received', 'damaged', 'yes', 'no']] = df[
    ['nominal', 'given', 'received', 'damaged', 'yes', 'no']].astype(int)

dictionary = {' край': '', ' автономная область': '', ' область': '', ' автономный округ':'',
              'Республика ': '', ' Республика': '', 'город ':'',
              ' \(Адыгея\)': '', ' \(Татарстан\)': '', ' \(Якутия\)':'',
             ' - Кузбасс':'', ' - Югра':''}
for key,val in dictionary.items():
    df['region'] = df['region'].str.replace(key, val)

print(np.unique(df['region']))
df.head()
```

```python
def latinize(s):
    """Latinsca"""
    a1 = 'абвгдежзиклмнопрстуфхцчшщэюяьъый'
    a2 = "a b v g d e zh z i k l m n o p r s t u f h ts ch sh sch e yu ya ' ' y y"
    a1 = a1 + a1.upper()
    a2 = a2.split(' ')
    a2 = a2 + [s.capitalize() for s in a2]
    t = {a1[i]: a2[i] for i in range(len(a1))}
    return ''.join([t[c] if c in t else c for c in s])
    
latinize('Цюцькин12')
```

```python
df['region_en'] = [latinize(s) for s in df.region.values]
df.head()
```

```python
df.to_csv('cleaned.csv', index=False)
```
