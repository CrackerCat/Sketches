# Equality

What % of total wealth would the top 10% of population own in various scenarios?

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gini(x):
    n = len(x)
    return 1-2/(n-1)*(n-np.dot(np.arange(1,n+1),sorted(x))/(np.sum(x)))
```

```python
n = 10000
mean_wealth = 100
sd_wealth = 38

wealth = np.maximum(0, mean_wealth + np.random.normal(size=n)*sd_wealth)
wealth = np.sort(wealth)

print('Owned by to 10\%:', np.sum(wealth[(n-n//10):]) / np.sum(wealth) * 100)
print('Owned by to 20\%:', np.sum(wealth[(n-n//5):]) / np.sum(wealth) * 100)
print('Average wealth of top 1\%:', np.sum(wealth[(n-n//100):]) / (n//100))
print('Average wealth of bottom 1\%:', np.sum(wealth[:n//100]) / (n//100))
print('Gini:', gini(wealth))

plt.figure(figsize=(6,5))
plt.plot(np.linspace(0,1,n), wealth);
plt.xlabel('Wealth percentile');
plt.ylabel('Wealth');
```

```python
# Same, but with exponential distribution

n = 10000
mean_wealth = 10
range_factor = 7

wealth = np.exp(np.random.uniform(size=n, low=0, high=range_factor))
wealth = wealth/np.mean(wealth)*mean_wealth
wealth = np.sort(wealth)

print('Owned by to 10\%:', np.sum(wealth[(n-n//10):]) / np.sum(wealth) * 100)
print('Average wealth of top 1\%:', np.sum(wealth[(n-n//100):]) / (n//100))
print('Average wealth of bottom 1\%:', np.sum(wealth[:n//100]) / (n//100))
print('Gini:', gini(wealth))

plt.figure(figsize=(6,5))
plt.plot(np.linspace(0,1,n), wealth);
plt.xlabel('Wealth percentile');
plt.ylabel('Wealth');
```

```python
plt.plot(np.linspace(0,1,n), np.cumsum(wealth) / np.sum(wealth));
plt.xlabel('Wealth percentile');
plt.ylabel('Cumulative wealth');
```
