### Power

Calculate x^n where n is integer

```python
import time
```

```python
def pow(x,n):
    m = abs(n)
    bag = 1 # Temporary bag
    res = 1 # Final destination
    i = 0   # Number of xs accumulated in the bag
    j = 0   # Number of xs accumulated in the res
    while i<m:
        if i==0:
            bag = x
            i = 1
        if (m-i)<2*i+j: # Jumping to 2i won't bring us over
            bag = bag*bag
            i = i*2
        else:
            j = j+i
            res = res*bag
            i = 0
            bag = 1
    if n>0:
        return res
    else:
        return 1/res

print(pow(3,3))
print(pow(2,-4))

tic = time.time()
print(pow(777,1000))
print(time.time()-tic)
```
