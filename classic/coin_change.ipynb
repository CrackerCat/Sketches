### Coins and change

How to give change with coins in a strange country with strange non-decimal coins

```python
coins = [1,2,5,7,11,23,51] # Set of currency
target = 27

coins.sort(reverse=True)
mem = {}

def solve(n):
    if n<=0:
        return []
    out = []
    for c in coins:
        if c==n:
            return [c]
        if c<n:
            if n-c in mem:
                part = mem[n-c]
            else:
                part = solve(n-c)
                mem[n-c] = part
            out.append((len(part)+1,[c]+part))
    #print(out)
    return out[0][1]

for i in range(100):
    print(i,solve(i))
```
