# Test dynamic matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
%matplotlib notebook
# This comment above is necessary; without it it doesn't render eagerly, so no animations

f1= plt.figure()
a1 = plt.axes()

n = 100
xy = np.random.rand(n,2)
maxTime = 5 # seconds
k = 0.01

tic = time.time()
print('start')
count = 0
while time.time()-tic < maxTime:
    #xy = xy+np.random.randn(n,2)/100
    for i in range(n):
        j = np.random.randint(0,n)
        xy[i,] += np.flip(xy[j,]-xy[i,])*[1,-1]*k
    xy = np.minimum(1,np.maximum(0,xy))
    
    if count==0:
        a1.plot(xy[:,0].transpose(),xy[:,1].transpose(),'.');
    else:
        a1.lines[0].set_xdata(xy[:,0])
        a1.lines[0].set_ydata(xy[:,1])
    f1.canvas.draw()
    count += 1
print('stop')
```
