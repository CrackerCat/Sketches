# Multivariate dynamics

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#%matplotlib notebook

nsteps = 5000

fig = plt.figure(figsize=(5,5));
#plt.xlim((0.5, 1.6))
#plt.ylim((0.5, 1.6))

x = 1
y = 1.2
z = 0.2
w = 0
history = np.zeros(shape=(1,3))
dt = 0.1

alpha = 1
K = 10
a = 1.3
b = 0.5
c = 1.6
d = 0.7

for i in range(nsteps):
    x += dt*(a*x*(1-x/K) - b*x/(alpha+x)*y)
    y += dt*(c*x/(alpha+x)*y - d*y - 0.15*y*w)
    z += dt*(0.15*y*w - z*0.02) # young bears
    w += dt*(z*0.02 - w*0.07)   # old bears
    history = np.append(history,np.array([[x,y,z]]),axis=0)
    
plt.plot(history);
    #fig.canvas.draw()
```
