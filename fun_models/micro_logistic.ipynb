# Minimalistic logistic map

```python
import matplotlib.pyplot as plt
```

```python
# A toy example of how recusrive lambda may work
f = lambda x: x if x>10 else f(2*x)
f(1)
```

```python
# Test the dynamics
f = lambda x,k,a: (x if k==0 else f(a*x*(1-x),k-1,a))
y = [0]*100
for i in range(100):
    y[i] = f(0.2,i,3.7)
    
plt.plot(list(range(100)),y);
```

```python
# Unpacked version (with a loop)
f = lambda x,k,a: (x if k==0 else f(a*x*(1-x),k-1,a))
n = 10000
y = [0]*n
for i in range(n):
    y[i] = f(0.21, 17 + (i % 19), 2.5+1.5*i/n)
    
plt.plot(list(range(n)),y,'.')
```

```python
# Minimalistic version (with list comprehension)
import matplotlib.pyplot as plt
f = lambda x,k,a: (x if k==0 else f(a*x*(1-x),k-1,a))
y = [f(.21,17 + i % 19,  2.5+1.5*i/10000) for i in range(10000)]
plt.plot(y,'.',markersize=1);
```
